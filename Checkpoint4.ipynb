```python
Hdf = df.drop(['id', 'date', 'bedrooms', 'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long', 'sqft_lot15'],axis = 1)      ### leaving only the well coorelated features
Hdf.head()        
```

```python
sns.pairplot(Hdf)    ### pairplot of our data
plt.show()
```

```python
plt.figure(figsize = (15,5))   ### number of bathrooms \ price
sns.pointplot(data = Hdf, x = 'bathrooms', y = 'price')
```

```python
from sklearn.model_selection import train_test_split    ### splitting data to train and test to train our model and test its accuracy
x=Hdf["sqft_living"].values[:,np.newaxis] 
y=Hdf["price"].values
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=50)
```

```python
from sklearn.linear_model import LinearRegression  ### using linear regression
from sklearn.metrics import mean_squared_error
from sklearn import metrics
model=LinearRegression()   
model.fit(x_train,y_train)
```

```python
plt.scatter(x,y,color="r")     ### plotting linear regression
plt.title("Linear Regression")
plt.ylabel("price")
plt.xlabel("sqft_living")
plt.plot(x,model.predict(x),color="k")
plt.show()                     ### after displaying the plot we can see some outliers
```

```python
Difference_N1 = y_test - model.predict(x_test)      ### we plot the difference between the predicted model and the results as we can see the model is working good so far
plt.plot((Difference_N1), marker='.', linestyle='-' )
```

```python
R2_1 = metrics.r2_score(y_test,model.predict(x_test))     ### we calculated main square error to see how the model is doing and we notice that half of the results are based on the model's input
print("R squared value 1 =", R2_1)
```
